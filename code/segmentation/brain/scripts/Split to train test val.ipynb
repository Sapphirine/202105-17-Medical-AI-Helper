```python
#!/usr/bin/env python 
# -*- coding:utf-8 -*-
# 
import os
import random
import shutil
from shutil import copy2
datadir_normal = "/home/gxt/study/CT/dataset/normal/"

all_data = os.listdir(datadir_normal)
num_all_data = len(all_data)
print( "num_all_data: " + str(num_all_data) )
index_list = list(range(num_all_data))
#print(index_list)
random.shuffle(index_list)
num = 0

trainDir = "/home/gxt/study/CT/dataset/normalTrain/"
if not os.path.exists(trainDir):
    os.mkdir(trainDir)
        
validDir = '/home/gxt/study/CT/dataset/normalVal/'
if not os.path.exists(validDir):
    os.mkdir(validDir)
        
testDir = '/home/gxt/study/CT/dataset/normalTest/'       
if not os.path.exists(testDir):
    os.mkdir(testDir)
        
for i in index_list:
    fileName = os.path.join(datadir_normal, all_data[i])
    if num < num_all_data*0.7:
        #print(str(fileName))
        copy2(fileName, trainDir)
    elif num>num_all_data*0.7 and num < num_all_data*0.8:
        #print(str(fileName))
        copy2(fileName, validDir)
    else:
        copy2(fileName, testDir)
    num += 1
```

```python
from glob import glob
from shutil import copy2
import os
train_files = []
mask_files = glob('/home/gxt/study/BrainSegmentation/BrainMRI/dataset/lgg-mri-segmentation/kaggle_3m/*/*_mask*')

for i in mask_files:
    train_files.append(i.replace('_mask',''))
    
trainImageFolder = '/home/gxt/study/BrainSegmentation/BrainMRI/dataset/splited/trainImage/'
trainMaskFolder = '/home/gxt/study/BrainSegmentation/BrainMRI/dataset/splited/trainMask/'

testImageFolder = '/home/gxt/study/BrainSegmentation/BrainMRI/dataset/splited/testImage/'
testMaskFolder = '/home/gxt/study/BrainSegmentation/BrainMRI/dataset/splited/testMask/'

valImageFolder = '/home/gxt/study/BrainSegmentation/BrainMRI/dataset/splited/valImage/'
valMaskFolder = '/home/gxt/study/BrainSegmentation/BrainMRI/dataset/splited/valMask/'

if not os.path.exists(trainImageFolder):
    os.mkdir(trainImageFolder)
if not os.path.exists(trainMaskFolder):
    os.mkdir(trainMaskFolder)    

if not os.path.exists(testImageFolder):
    os.mkdir(testImageFolder)
if not os.path.exists(testMaskFolder):
    os.mkdir(testMaskFolder)    

if not os.path.exists(valImageFolder):
    os.mkdir(valImageFolder)
if not os.path.exists(valMaskFolder):
    os.mkdir(valMaskFolder)    
    
num = 0
total = len(train_files)

for img in train_files:
    if num < total*0.7:
        copy2(img, trainImageFolder)
    if num > total*0.7 and num < total*0.8:
        copy2(img, valImageFolder)
    if num > total*0.8:
        copy2(img, testImageFolder)
    num += 1

num = 0
for mask in mask_files:
    if num < total*0.7:
        copy2(mask, trainMaskFolder)
    if num > total*0.7 and num < total*0.8:
        copy2(mask, valMaskFolder)
    if num > total*0.8:
        copy2(mask, testMaskFolder)
    num += 1
print(num)
```
